## HomeWork 1-1. -不等式制約条件付き最適化
制約条件
$$
c_1(x_1, x_2) = x_1 + 3x_2-8 \leq 0
$$
$$
c_2(x_1, x_2) = x_1^2 + x_2^2 + 2x_1 - 2x_2 -3 \leq 0
$$
のもとで，
$$f(x_1,x_2)=x_1^2 +2x_1x_2+ 2x_2^2 -10x_1 -12x_2
$$
を最小化する$x_1, x_2$を求める．

$f$のHessian $H$ について，
$$
H = \left(
\begin{matrix} 
2 & 2 \\ 
2 & 4 
\end{matrix} 
\right)
$$
であり，その固有値は$3\pm\sqrt{5} > 0$である．従って，$f$は凸関数である．

$f$の凸性より，内点法を用いて数値計算を行う．新たに目的関数 
$$
\bar{f}_{p,q} = f + \frac{1}{pc_1^2} + \frac{1}{qc_2^2}
$$
を導入して，制約なし最適化問題として扱うことができる．
$p,q$が十分大きい際に $\bar{f}$を最小にする$x_1,x_2$は元の目的関数の最適解である．

```julia
using Plots, LinearAlgebra
```

```julia
f(x₁,x₂) = x₁^2 + 2x₁*x₂ + 2x₂^2 - 10x₁ - 12x₂
c₁(x₁,x₂) = x₁ + 3x₂ - 8
c₂(x₁,x₂) = x₁^2 + x₂^2 + 2x₁ - 2x₂ - 3
f_bar(x₁,x₂, p,q) = f(x₁,x₂) + (p*c₁(x₁,x₂)^2)^-1 + (q*c₂(x₁,x₂)*2)^-1
```

最適化にあたっては準ニュートン法を使用する．すなわち計算の１ステップでの修正量$\Delta x_k$について，
$$
\Delta x_k = -H_k^{-1}F(x_k)
$$
となるようなへシアンの近似行列 $H_k$を計算する．ただし，$F$は目的関数の勾配ベクトルである．この更新式は，
$$ s_k = \Delta x_k = x_{k+1} - x_k, \qquad y_k = F(x_{k+1})-F(x_k)$$
として，ブロイデン・フレッチャー・ゴールドファーブ・シャンノ（BFGS)法
$$
H_{k+1} = H_k + \frac{y_ky_k^T}{y_k^Ts_k} - \frac{H_ks_ks_k^TH_k}{s_k^TH_ks_k}
$$
を用いて計算できる．

```julia
∂f∂x₁(x₁, x₂) = 2x₁ + 2x₂ - 10
∂f∂x₂(x₁, x₂) = 2x₁ + 4x₂ - 12
∂c₁∂x₁(x₁, x₂) = 1
∂c₁∂x₂(x₁, x₂) = 3
∂c₂∂x₁(x₁, x₂) = 2x₁ + 2
∂c₂∂x₂(x₁, x₂) = 2x₂ - 2
∂f₁(x₁,x₂, p,q) = ∂f∂x₁(x₁, x₂) -2(p*c₁(x₁, x₂)^3)*∂c₁∂x₁(x₁, x₂) -2(q*c₂(x₁, x₂)^3)*∂c₂∂x₁(x₁, x₂)
∂f₂(x₁,x₂, p,q) = ∂f∂x₂(x₁, x₂) -2(p*c₂(x₁, x₂)^3)*∂c₁∂x₁(x₁, x₂) -2(q*c₂(x₁, x₂)^3)*∂c₂∂x₂(x₁, x₂)
F(x₁, x₂, p,q) = [∂f₁(x₁,x₂, p,q), ∂f₂(x₁,x₂, p,q)]
Δx_k(H_k, x₁, x₂, p,q) = -inv(H_k)*F(x₁, x₂, p,q)
H_knew(H,y,s) = H + y*transpose(y)/(transpose(y)*s) - H*s*transpose(s)*H/(transpose(s)*H*s)
```

実際に$x$を更新する際は適切なスカラー$\alpha$をかけて修正幅を修正する．そのような$\alpha$を直線探索アルゴリズムで探索する．
$$
x_{k+1} = x_k + \alpha \Delta x_k
$$

```julia
function BracketingSearch(ϕ, dim::Int, p, q)
    k = 0
    α = zeros(dim)
    h = 1e-6
    ste = h*ones(dim)
    res = ϕ(α[1], α[2], p,q)
    for i in 1:1e10
        if ϕ(α[1], α[2], p,q) > res
            break
        end
        α .+= h
    end
#     println(α)
    return α
end
```

```julia
function HW_QuasiNewton(x₀, H₀, p,q)
    ε = 1e-5     # 収束判定
    k = 0
    x = x₀
    H = H₀
    x_new = zeros(2)
    Δx = zeros(2)
    α = zeros(2)
    for i in 1:1e4
        if norm(F(x[1], x[2], p,q)) < ε
            break
        end
        Δx = Δx_k(H, x[1], x[2], p,q)
        α = BracketingSearch(f_bar, 2, p,q)
        x_new = x + α.*Δx
        H = H_knew(H, F(x_new[1], x_new[2], p,q)-F(x[1], x[2], p,q), x_new-x)
        x = x_new
    end
    return x, H
end
```

```julia
function main()
#     x = zeros(2)
    x = [0., 0.]
    H = [1. 0.; 0. 1.]
    p=1e-12
    q=1e-12
    ε = 1e-5     # 収束判定
    while p < 1e-1
        x, H = HW_QuasiNewton(x, H, p,q)
        println(p)
        print(x)
        println(f(x[1], x[2]))
        if norm(x) < ε
            break
        end
        p*=1.1
        q*=1.1
#         if p<1
#             p *= 2
#             q *= 2
#         else
#             p+=10
#             q+=10
#         end
    end
    
    return x
end
```
